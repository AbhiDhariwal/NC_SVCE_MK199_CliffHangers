```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img_real_1 = cv2.imread("p1.jpg")
img_r_1 = cv2.imread("p1.jpg",cv2.IMREAD_GRAYSCALE)
img_1 = cv2.GaussianBlur(img_r_1,(3,3),0)

lap_1 = cv2.Laplacian(img_1, cv2.CV_64F, ksize=3)
lap_1 = np.uint8(np.absolute(lap_1))
_, mask_1 = cv2.threshold(lap_1, 50,255, cv2.THRESH_BINARY)

img_real_2 = cv2.imread("p2.jpg")
img_r_2 = cv2.imread("p2.jpg",cv2.IMREAD_GRAYSCALE)
img_2 = cv2.GaussianBlur(img_r_2,(3,3),0)

lap_2 = cv2.Laplacian(img_2, cv2.CV_64F, ksize=3)
lap_2 = np.uint8(np.absolute(lap_2))
_, mask_2 = cv2.threshold(lap_2, 50,255, cv2.THRESH_BINARY)
pixel_1 = mask_1.reshape(-1)
pixel_1 = list(pixel_1)
n_1 = pixel_1.count(255)
m_1  = pixel_1.count(0)
ratio_1 = n_1/m_1

pixel_2 = mask_2.reshape(-1)
pixel_2 = list(pixel_2)
n_2 = pixel_2.count(255)
m_2  = pixel_2.count(0)
ratio_2 = n_2/m_2


fig, axs = plt.subplots(2, 2)
axs[0, 0].imshow(img_real_1)
axs[0, 1].imshow(img_real_2)
axs[1, 0].imshow(mask_1)
axs[1, 0].set_title(ratio_1)
axs[1, 1].imshow(mask_2)
axs[1, 1].set_title(ratio_2)
```
